Read the data. Call the loaded data `college`. Make sure that you have the directory set to the correct location for the data.

```python
college = pd.read_csv('datasets/College.csv')
```

i. Use the `describe()` function to produce a numerical summary of the variables in the data set.

```python
college.describe()
```

ii. Use `sns.pairplot` function to produce a scatterplot matrix of the first ten columns or variables of the data

```python
sns.pairplot(college.iloc[:, :10]);
```

iii. Use `sns.boxplot` function to produce side-by-side boxplots of Outstate versus Private.

```python
sns.boxplot(x='Private', y='Outstate', data=college);
```

iv. Create a new qualitative variable, called `Elite`, by binning the `Top10perc` variable. We are going to divide universities into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50 %.

```python
college['Elite'] = college.Top10perc > 50
```

Use the `sum()` function to see how many elite universities there are. 

```python
sum(college.Elite)
```

Now use the `sns.boxplot` function to produce side-by-side boxplots of Outstate versus Elite.

```python
sns.boxplot(x='Elite', y='Outstate', data=college);
```

v. Use the `sns.distplot` function to produce some histograms with differing numbers of bins for a few of the quantitative variables

```python
ax = sns.distplot(college.Books, bins=50, kde=False)
ax.set_title('Book cost');
```

```python
ax = sns.distplot(college['Room.Board'], bins=50, kde=False)
ax.set_title('Room and board cost');
```

```python
ax = sns.distplot(college['PhD'], bins=50, kde=False)
ax.set_title('Percent of faculty with PhD');
```
